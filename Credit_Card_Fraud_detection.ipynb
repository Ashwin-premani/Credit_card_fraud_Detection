```python
# Importing essential libraries for data manipulation, visualization, and model building
import pandas as pd
```

```python
# Reading the dataset to load transaction records for credit card fraud detection analysis
data = pd.read_csv('creditcard.csv')
```

```python
from imblearn.over_sampling import SMOTE
# Dropping irrelevant columns to reduce noise and improve model performance
X = data.drop('Class', axis=1)
y = data['Class']
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)
```

```python
data['Hour'] = data['Time'] // 3600 % 24
```

```python
data.duplicated().sum()
data.drop_duplicates(inplace=True)
```

```python
# Importing essential libraries for data manipulation, visualization, and model building
import numpy as np
q_low = data['Amount'].quantile(0.01)
q_hi  = data['Amount'].quantile(0.99)
data = data[(data['Amount'] >= q_low) & (data['Amount'] <= q_hi)]
```

```python
# Importing essential libraries for data manipulation, visualization, and model building
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data['Amount_scaled'] = scaler.fit_transform(data[['Amount']])
```

```python
# Dropping irrelevant columns to reduce noise and improve model performance
data.drop(['Time', 'Amount'], axis=1, inplace=True)
```

```python
# Splitting the dataset into training and testing sets to evaluate model performance on unseen data
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
# Generating a confusion matrix to visualize true vs predicted classifications 
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Dropping irrelevant columns to reduce noise and improve model performance
X = data.drop('Class', axis=1)
y = data['Class']

# Splitting the dataset into training and testing sets to evaluate model performance on unseen data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
model.fit(X_train_res, y_train_res)
```

```python
# Using the trained model to predict fraud on the test dataset
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

# Generating a confusion matrix to visualize true vs predicted classifications
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
# Getting precision, recall, f1-score to deeply understand the model's strengths and weaknesses
print("\nClassification Report:\n", classification_report(y_test, y_pred, digits=4))
print("ROC-AUC Score:", roc_auc_score(y_test, y_proba))
```

```python
# Importing essential libraries for data manipulation, visualization, and model building
import joblib
joblib.dump(model, "credit_fraud_rf_model.pkl")
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train_res, y_train_res)
# Using the trained model to predict fraud on the test dataset
y_pred_lr = lr_model.predict(X_test)
print("Logistic Regression:")
# Getting precision, recall, f1-score to deeply understand the model's strengths and weaknesses
print(classification_report(y_test, y_pred_lr))
# Using the trained model to predict fraud on the test dataset
print("ROC-AUC Score:", roc_auc_score(y_test, lr_model.predict_proba(X_test)[:, 1]))
```
